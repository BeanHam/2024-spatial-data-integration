```python
import numpy as np
import pandas as pd
import scienceplots
import seaborn as sns
import matplotlib.pyplot as plt
```

### join

```python
models=['llama3', 'mistral', '4o-mini', '4o', 'best\nheuristic']*4
methods=['zero-shot']*10+['few-shot']*10
modes=(['no-heuristic']*5+['with-heuristic']*5)*2
acc=[0.474,0.601,0.445,0.496,0.943,0.529,0.601,0.813,0.914,0.943,0.478,0.516,0.514,0.634,0.943,0.617,0.519,0.880,0.927,0.943]
vis_data=pd.DataFrame(np.c_[models, methods, modes, acc],columns=['model', 'zero/few-shot', 'no/with-heuristics', 'acc'])
vis_data = vis_data.astype({"acc": float})
```

```python
plt.style.use('science')
plt.rc('text', usetex=False)
plot=sns.catplot(vis_data, kind="bar",x="model", y="acc", col="zero/few-shot", hue='no/with-heuristics',height=3.5, aspect=1.2)
plot.set_xlabels("")
plot.set_xticklabels(fontsize=13)
plot.set_ylabels("Spatial Join Accuracy", fontsize=14)
plot.legend.set_title("")
plot.set_titles(col_template="{col_name}", size=16)
sns.move_legend(plot, "upper left", bbox_to_anchor=(.08, .9))
plt.savefig('join_task_no_finetuning.png', bbox_inches='tight',pad_inches=0, dpi=600)
pass
```
